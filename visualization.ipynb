```python
import matplotlib.pyplot as plt


students_names=["sanjay","Rahul","Karan","Wasim","Ramesh","Ajay","Sartaj","Priya"]

students_marks=[35,50,20,45,25,40,25,40]

mark_perc = []

for x in students_marks:
    res = (x / 50) * 100
    mark_perc.append(res)
    
def line_chart_of_students_and_marks():
    plt.plot(students_names,students_marks, marker = "x" )
    plt.title("Student Marks Graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Marks")
    plt.show()
    
line_chart_of_students_and_marks()
```

```python
def percentage_bar_chart():
    plt.bar(students_names, mark_perc, )
    plt.title("Students Percentage Graph")
    plt.xlabel("Students Names")
    plt.ylabel("Students Percentage")
    plt.show()
    
percentage_bar_chart()
```
